# Data Visualizations

### Scatter Plot
        -Analyze relationship between variables

```python
plt.figure()
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species'
)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

### Histogram
        -Examine data distribution

```python
plt.figure()
plt.hist(df['petal_length'], bins=20)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

### Box Plot
        -Detect outliers & spread

```python
plt.figure()
sns.boxplot(data=df, x='species', y='sepal_length')
plt.title('Sepal Length by Species')
plt.show()
```

```python
--Final Results & Insights (For Report):
The Iris dataset contains 150 rows and 5 columns.
No missing values were found.
Scatter plots show clear relationships between features.
Histograms highlight data distribution.
Box plots help identify spread and potential outliers.
Species-wise separation is clearly visible in visualizations.
```
